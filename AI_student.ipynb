EDA (Exploratory DATA Analiysis)

```python
%pip install seaborn
import seaborn as sns
```

```python
#Univariate Analiysis
df = pd.read_csv("ai_student_impact_dataset.csv")
```

```python
df.head()
```

1.Categorical Data

```python
#countplot

sns.countplot(df['Major_Category'])
df['Major_Category'].value_counts()
```

```python
sns.countplot(df['Prompt_Engineering_Skill'])
df['Prompt_Engineering_Skill'].value_counts()
```

```python
df['Prompt_Engineering_Skill'].value_counts().plot(kind='pie', autopct='%.2f')
```
